```python
from pinecone import Pinecone
pc = Pinecone()
```

```python
'''
Create and Delete Indexes
'''
from pinecone import ServerlessSpec
index_name = 'langchain'
if index_name not in pc.list_indexes().names():
    print(f"Creating Index: {index_name}")
    pc.create_index(
        name = index_name,
        dimension = 1536,
        metric = "cosine",
        spec = ServerlessSpec(
            cloud = "aws",
            region = "us-east-1"
        )
    )
else:
    print(f"Index name already exists")
        
    
```

```python
index_name = 'langchain'
if index_name in pc.list_indexes().names():
    print(f"Deleting index {index_name} ...")
    pc.delete_index(index_name)
    print('Done')
else:
    print(f"Index name {index_name} does not exist")
```

```python
'''
Working with Vectors

inserting vectors
updating vectors
fetching vectors
delete vectors
'''
import random
vectors = [[random.random() for _ in range(1536)] for v in range(5)]
# print(vectors)
ids = list('abcde')

index_name = 'langchain'
index = pc.Index(index_name)

index.upsert(vectors=zip(ids, vectors))
```

```python
'''
Working with Vectors

inserting vectors
updating vectors
fetching vectors
delete vectors
'''
import random
vectors = [[random.random() for _ in range(1536)] for v in range(5)]
# print(vectors)
ids = list('abcde')

index_name = 'langchain'
index = pc.Index(index_name)
print(f"Index -- {index}")

index.upsert(vectors=zip(ids, vectors))

# query
query_vector = [random.random() for _ in range(1536)]

index.query(
    vector = query_vector,
    top_k = 3,
    include_values = False
)
```

```python
'''
Working with Namespaces

inserting vectors
updating vectors
fetching vectors
delete vectors
'''
import random
vectors = [[random.random() for _ in range(1536)] for v in range(5)]
# print(vectors)
ids = list('abcde')

index_name = 'langchain'
index = pc.Index(index_name)
print(f"Index -- {index}")

index.upsert(vectors=zip(ids, vectors))

vectors = [[random.random() for _ in range(1536)] for v in range(3)]
ids = list('xyz')
index.upsert(vectors=zip(ids, vectors), namespace='first-namespace')
```

```python
vectors = [[random.random() for _ in range(1536)] for v in range(2)]
ids = list('qp')
index.upsert(vectors=zip(ids, vectors), namespace='second-namespace')
```
